# BLIP 2 Demo

Use BLIP 2 through huggingface's `transformers` to caption images and answer questions about them.

Based mainly on [this excellent example notebook](https://github.com/NielsRogge/Transformers-Tutorials/blob/master/BLIP-2/Chat_with_BLIP_2_%5Bint8_bitsandbytes%5D.ipynb)

```python
import requests
import time

import torch

from PIL import Image
from transformers import AutoProcessor, Blip2ForConditionalGeneration
```

## Load an image
Can be a URL or a local image file

```python
# Uncomment to use a URL
url = 'https://www.noisebridge.net/images/0/09/VideoEditing.jpg' 
image = Image.open(requests.get(url, stream=True).raw).convert('RGB')

# Uncomment to use a local image file
# file = 'some local file path'
# image = Image.open(file).convert('RGB')  

display(image.reduce(3))
```

## Load a BLIP 2 model

Full checkpoint zoo is [here](https://huggingface.co/models?sort=downloads&search=blip2). Using `bitsandbytes`, load the model weights as 8bit values to save memory.

Also loads an appropriate preprocessor for fitting input image and text data into the model. 

```python
checkpoint = 'Salesforce/blip2-flan-t5-xxl'

processor = AutoProcessor.from_pretrained(checkpoint)
model = Blip2ForConditionalGeneration.from_pretrained(checkpoint, device_map='auto', load_in_8bit=True)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

**Get a parameter count**

Just for fun

```python
def get_n_params(model):
    pp=0
    for p in list(model.parameters()):
        nn=1
        for s in list(p.size()):
            nn = nn*s
        pp += nn
    return pp

n_params = get_n_params(model)
print(f'Loaded model has {n_params:,} parameters')
```

## Caption the input image

Caption content generation can be influenced with the arguments to `model.generate`

```python
def caption_image(image, model, processor, device, **kwargs) -> str:
    """Return an image caption given a BLIP 2 model.
    
    If desired, generate kwargs (max_new_tokens, temperature, do_sample, 
    num_beams, etc.) can be passed in as `kwargs`.
    
    """
    if len(kwargs) == 0:
        # Default settings
         kwargs = {
            'max_new_tokens': 500,
            'temperature': 1.1,
            'do_sample': True,
            'num_beams': 4}
        
    inputs = processor(
        image, 
        return_tensors="pt").to(device, torch.float16)

    generated_ids = model.generate(**inputs, **kwargs)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
    return generated_text
```

```python
t0 = time.time()
generated_text = caption_image(
    image, model, processor, device,
    max_new_tokens=50, temperature=1.1, do_sample=True, num_beams=4)
t1 = time.time()
print(generated_text)
print(f'Elapsed time: {t1-t0:.2f} s')
```

## Visual question answering
Answer questions about the input image! Each question is independent of previous questions.

```python
def answer_question(image, question, model, processor, device, **kwargs):
    """Answer a question about an image.
    
    If desired, generate kwargs (max_new_tokens, temperature, do_sample, 
    num_beams, etc.) can be passed in as `kwargs`.
    
    """
    if len(kwargs) == 0:
        # Default settings
         kwargs = {
            'max_new_tokens': 500,
            'temperature': 1.,
            'do_sample': True,
            'num_beams': 4}

    prompt = f'Question: {question} Answer:'
    
    inputs = processor(image, text=prompt, return_tensors="pt").to(device, torch.float16)

    generated_ids = model.generate(**inputs, **kwargs)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
    return generated_text
    
```

```python
question1 = 'How many faces are here?'
question2 = 'How many computer screens are here?'
question3 = "What color is the shirt of the person on the right?"

for q in [question1, question2, question3]:
    t0 = time.time()
    a = answer_question(image, q, model, processor, device)
    t1 = time.time()
    print(a)
    print(f'Elapsed time: {t1-t0:.2f} s')
```

## Chat mode
Concatenate together questions and answers to form longer contexts in a Q&A-style format. End a session by typing "!end".

The max context window length is 512 tokens.

```python
def qna(image, model, processor, device, **kwargs):
    """Hold a Q&A-style session regarding the input image.
    
    If desired, generate kwargs (max_new_tokens, temperature, do_sample, 
    num_beams, etc.) can be passed in as `kwargs`.
    
    """
    if len(kwargs) == 0:
        # Default settings
         kwargs = {
            'max_new_tokens': 500,
            'temperature': 1.,
            'do_sample': True,
            'num_beams': 4}
            
    context = []
    while True:
        description = '> '
        question = input(description)
        if question.strip().lower() == '!end':
            break
        prompt = ' '.join([f'Question: {c[0]} Answer: {c[1]}' for c in context]) + ' Question: ' + question + ' Answer: '
        inputs = processor(image, text=prompt, return_tensors='pt').to(device, torch.float16)
        generated_ids = model.generate(**inputs, **kwargs)
        generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
        print(generated_text)
        context.append((question, generated_text))
    return
```

```python
qna(image, model, processor, device)
```
